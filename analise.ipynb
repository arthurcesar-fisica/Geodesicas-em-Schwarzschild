# Geodésica de Scharzschild

### Análise do Problema

Queremos realizar uma simulação visual de como fica a trajetória de um fóton quando esse se encontra próximo de um buraco negro. 
Para esse problema, precisamos compreender como é dado a solução de Scharzschild. Quando solucionamos as equações de campo de Einstein, no vácuo e com simetria esférica, o teorema de Birkhoff garante que a única solução possível que não explode no infitino é a métrica de Scharzschild. Quando consideramos o sistema estático, então a métrica é independente do tempo, se é esfericamente simétrico o ângulo não importa(*).
